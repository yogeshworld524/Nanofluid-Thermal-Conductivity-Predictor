```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor

# Load Dataset
df = pd.read_csv("nanofluid_data.csv")

# Encode categorical variables
particle_encoder = LabelEncoder()
basefluid_encoder = LabelEncoder()

df["Particle"] = particle_encoder.fit_transform(df["Particle"])
df["BaseFluid"] = basefluid_encoder.fit_transform(df["BaseFluid"])

# Features and Target
X = df[["Temperature", "Concentration", "Particle", "BaseFluid"]]
y = df["ThermalConductivity"]

# Train Model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Predict
temperature = 50
concentration = 0.04
particle = particle_encoder.transform(["GNP"])[0]
basefluid = basefluid_encoder.transform(["Water"])[0]

prediction = model.predict([[temperature, concentration, particle, basefluid]])

print(f"Predicted Thermal Conductivity = {prediction[0]:.4f} W/mK")
```
